```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

### PARTE 1 - Carga de datos y preparacion

```python
input_file = "banking.csv"
df = pd.read_csv(input_file, header=0)
df.head()
```

```python
print("Training Data Frame")
print(pd.isnull(df).sum()) 
```

```python
X = df.loc[:, df.columns != 'y']
X
```

```python
Y = df['y'].values
Y
```

```python
categorical_columns = X.select_dtypes(include=['object']).columns.tolist()
encoder = OneHotEncoder(drop='first', sparse=False) 
X_encoded = pd.DataFrame(encoder.fit_transform(X[categorical_columns]))
X_encoded.columns = encoder.get_feature_names_out(categorical_columns)
X = pd.concat([X.drop(categorical_columns, axis=1), X_encoded], axis=1)
```

### Parte 2: entrenamiento y testing

```python
train_X, test_X, train_y, test_y = train_test_split(X, Y, test_size=0.30, random_state=
0, shuffle=True)
```

```python
lr = LogisticRegression()
lr = lr.fit(train_X, train_y)
```

### Parte 3: Evaluacion

```python
y_pred = lr.predict(test_X)
print("Predicted vs Expected")
print(y_pred)
print(test_y)
```

```python
print(classification_report(test_y, y_pred, digits=3))
```

```python
print(confusion_matrix(test_y, y_pred))
```
