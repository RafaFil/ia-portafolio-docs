Discriminante Lineal utilizando Python y scikit-learn

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
input_file = "sports_Training.csv"
d_training = pd.read_csv(input_file, header=0)
d_training
```

```python
input_file = "sports_Scoring.csv"
d_scoring = pd.read_csv(input_file, header=0)
d_scoring
```

```python
data = d_training[(d_training['CapacidadDecision'] >= 3) & (d_training['CapacidadDecision'] <= 100)]
data
```

```python
data_scor_filtred = d_scoring[(d_scoring['CapacidadDecision'] >= 3) & (d_scoring['CapacidadDecision'] <= 100)]
data_scor_filtred
```

```python
X = data.loc[:, data.columns != 'DeportePrimario']
X
```

```python
y = data['DeportePrimario'].values
y
```

### Parte 2: Entrenamiento

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X,y)
```

### Parte 3: EvaluaciÃ³n

```python
y_pred = lda.predict(data_scor_filtred)
print("Predicted")
pred = pd.DataFrame(y_pred)
pred.columns = ["DeportePrimario"]
pred
```
